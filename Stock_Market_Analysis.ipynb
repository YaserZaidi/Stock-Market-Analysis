```python
import datetime
pd.set_option('mode.chained_assignment', None)
```

```python
today = date.today()
```

```python
d1 = today.strftime("%Y-%m-%d")
```

```python
end_date = d1
```

```python
d2 = date.today() - timedelta(days = 365)
d2 = d2.strftime('%Y-%m-%d')
start_date = d2
```

```python
data = yf.download('GOOG',
                      start = start_date,
                      end = end_date,
                      progress = True)
```

```python
data
```

```python
data["Date"] = data.index
data = data[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
data.reset_index(drop = True, inplace = True)
print(data.head(12))
```

```python
figure = go.Figure(data =[go.Candlestick(x=data["Date"],
                                            open=data["Open"], high=data["High"],
                                            low=data["Low"], close=data["Close"])])
figure.update_layout(title= "Google Price Analysis", xaxis_rangeslider_visible = True)
figure.show()
```

```python
figure = px.bar(data, x = 'Date', y = 'Close')
figure.update_layout(xaxis_rangeslider_visible = True)
figure.show()
```

```python
figure = px.line(data, x = 'Date', y = 'Close', title = "Stock Market Analysis using Line Chart")
figure.update_layout(xaxis_rangeslider_visible = True)
figure.show()
```

```python
figure = px.line(data, x='Date', y='Close', 
                 title='Stock Market Analysis with Time Period Selectors')

figure.update_xaxes(
    rangeselector=dict(
        buttons=list([
            dict(count = 1, label = '1 min', step = "minute", stepmode = "todate"),
            dict(count = 3, label = '3 min', step = "minute", stepmode = "todate"),
            dict(count = 5, label = '5 min', step = "minute", stepmode = "todate"),
            dict(count = 10, label = '10 min', step = "minute", stepmode = "todate"),
            dict(count = 15, label = '15 min', step = "minute", stepmode = "todate"),
            dict(count = 30, label = '30 min', step = "minute", stepmode = "todate"),
            dict(count = 1, label = '1 hr', step = "hour", stepmode = "todate"),
            dict(count = 2, label = '2 hr', step = "hour", stepmode = "todate"),
            dict(count = 3, label = '3 hr', step = "hour", stepmode = "todate"),
            dict(count = 4, label = '4 hr', step = "hour", stepmode = "todate"),
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=3, label="3m", step="month", stepmode="backward"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)
```
